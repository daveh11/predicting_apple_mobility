```python
super_rows = [
    ["Pegasus", "True"],
    ["Unicorn", "False"]
]
sub_rows = [
    ["Alicorn", "23.1", "True"],
    ["basilisk", "120.433", "False"],
    ["ChiMera", "101.8", "True"],
    ["Nemean lion", "2", "False"],
    ["Griffin", "65.82,465.21", "False"],
    ["0rthros", "52.01", "False"],
    ["Cockatrice", "nan", "True"],
]

rows = super_rows + sub_rows

# Add ingestion code here


from dataclasses import dataclass
import numpy as np

@dataclass
class ParentClass:
    str_var: str
    bool_var: str
        
    def format_string(self): 
        return self.str_var.lower().capitalize()
    
    def format_bool(self):
        return True if self.bool_var.lower() == 'true' else False

    def display_formated(self):
        str_var_format = ParentClass.format_string(self)
        bool_var_format = ParentClass.format_bool(self)
        print(f'Name is: { str_var_format }\nBinary is: { bool_var_format } ' )

# instance of the "ParentClass"
'''parent = ParentClass("Pegasus", "True")

# stored values
print(parent.str_var)
print(parent.bool_var)

# display formated 
parent.display_formated()'''

@dataclass
class ChildClass(ParentClass):
    num_var: str
    
    def display_formated(self):
        str_var_format = ParentClass.format_string(self)
        bool_var_format = ParentClass.format_bool(self) 
        try:
            num_var_format = float(self.num_var)
        except:
            num_var_format = np.nan
            
        print(f'Name is: { str_var_format }\nBinary is: {bool_var_format} \nValue is: {num_var_format}' )

# instance of the "ParentClass"
#child = ChildClass(str_var="Alicorn", num_var="-", bool_var="True")

@dataclass
class factory:
    data:list

    def display_object(self):
        if len(self.data)==2:
            parent=ParentClass(self.data[0],self.data[1])
            parent.display_formated()
            print()
        else:
            child=ChildClass(str_var=self.data[0], num_var=self.data[1], bool_var=self.data[2])
            child.display_formated()
            print()

# stored values
'''print(child.str_var)
print(child.bool_var)
print(child.num_var)

# display formated 
child.display_formated()'''

# iterating through the 'rows' and creating objects
for i in rows:
    object_ = factory(i)
    object_.display_object()
```
